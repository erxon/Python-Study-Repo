```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Date": pd.date_range(start="2024-01-01", periods=30),
    "Close": [
        100, 102, 101, 105, 107, 106, 108, 110, 112, 111,
        115, 117, 116, 118, 120, 123, 122, 125, 128, 130,
        129, 131, 134, 136, 138, 140, 142, 145, 147, 150
    ]
}
```

```python
df = pd.DataFrame(data)
```

```python
df["SMA_5"] = df["Close"].rolling(window=5).mean()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df["Date"], df["Close"], label="Close Price", color="blue")
plt.plot(df["Date"], df["SMA_5"], label="5-Day SMA", color="orange", linewidth=2)
plt.title("Simple Moving Average (SMA) Example")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```
